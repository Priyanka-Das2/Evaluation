```python
import pandas as pd
```

```python
a  = pd.read_csv('anom_order.out', sep = '_', header = None, names = ['ed',' ','eth','File', 'anom','Date'])
a['Date'] = a['Date'].str.split('.').str[0]
a
```

```python
a['Date'] = pd.to_datetime(a['Date'], unit = 'ns')
a1 = pd.DataFrame(a.File.value_counts())
a11 = a1.rename(columns = {'File' : 'No_of_files'})
a11.index.name = 'File'
a11.head(5)
a11.to_csv('no_of_files_created.csv')
a11
```

```python
a['new_date'] = [i.date() for i in a['Date']]
a['new_time'] = [j.time() for j in a['Date']]
```
