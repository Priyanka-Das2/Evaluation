```python
trade_data = pd.DataFrame(data)

trade_data = trade_data.transpose()
trade_data = trade_data[['last','bid','ask']]
trade_data = pd.DataFrame(trade_data)
trade_data = trade_data.reset_index()
trade_data['pair'] = trade_data['index']
trade_data =  trade_data[['pair','last','bid','ask' ]]
trade_data['price'] = trade_data['last'] 
trade_data =  trade_data[['pair','price','bid','ask' ]]
trade_data 
```

```python
l = []
for j in trade_data['pair']:
    l.append(j.split('_')[1])
trade_data['crosscoin'] = l

trade_data['price'] = trade_data['price'].astype(float)
grouped = trade_data.groupby(['crosscoin']).agg({'price': "mean"})
grouped = grouped.sort_values('price', ascending = False)
trade_data['crosscoin'].unique()
grouped = grouped.drop(['ETH'])
grouped1 =  grouped.sort_values('price')
bid_group = trade_data.groupby(['crosscoin']).agg({'price': "mean"})
```
